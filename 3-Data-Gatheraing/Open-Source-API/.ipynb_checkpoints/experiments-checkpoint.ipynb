## API
`api is one kind of data pipeline whihe help to interact to different software and most of the time they share the data in JSON format.`

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["TMDB_API_KEY"]=os.getenv("TMDB_API_KEY")
tmdb_api_key = os.getenv("TMDB_API_KEY")
```

## Api link of data
`https://api.themoviedb.org/3/movie/top_rated?api_key=tmdb_api_key&page=1`

```python
import pandas as pd
import requests
```

```python
response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key={tmdb_api_key}&page=1')
```

```python
temp_df = pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
```

```python
temp_df.head(5)
```

```python
df = pd.DataFrame()  # Initialize an empty DataFrame

for i in range(1, 496):
    response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key={tmdb_api_key}&page={i}')
    temp_df = pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
    df = pd.concat([df, temp_df], ignore_index=True)  # Use pd.concat instead of append
```

```python
df.sample(4)
```

```python
from datetime import datetime

# Generate filename with today's date
file_name = f"top_rated_movies_{datetime.today().strftime('%Y-%m-%d')}.csv"

# Save the DataFrame to a CSV file
df.to_csv(file_name, index=False)

print(f"DataFrame saved to {file_name}")
```

```python
import requests
import pandas as pd

# API details
url = "https://imdb236.p.rapidapi.com/imdb/upcoming-releases"
querystring = {"countryCode": "US", "type": "MOVIE"}
headers = {
    "x-rapidapi-key": "Your_api_key",  # Replace with your RapidAPI key
    "x-rapidapi-host": "imdb236.p.rapidapi.com"
}

# Send the request
response = requests.get(url, headers=headers, params=querystring)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()

    # Print the response to understand its structure
    import json
    print(json.dumps(data, indent=4))

    # Adjust the key based on the actual structure of the JSON
    if 'upcoming_releases' in data:  # Replace 'upcoming_releases' with the correct key
        df = pd.DataFrame(data['upcoming_releases'])[['title', 'release_date']]
        # print(df)
    else:
        print("The expected key was not found in the response.")
else:
    print(f"Failed to fetch data: {response.status_code}")
```
