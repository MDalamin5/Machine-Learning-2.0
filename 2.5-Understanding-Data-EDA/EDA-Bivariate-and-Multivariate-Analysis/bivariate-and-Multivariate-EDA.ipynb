```python
import pandas as pd
import seaborn as sns

titanic = pd.read_csv("https://raw.githubusercontent.com/MDalamin5/Machine-Learning-2.0/refs/heads/main/2.5-Understanding-Data-EDA/Basics-Question-To-Data/train.csv")
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
flights = sns.load_dataset('flights')
flights.sample(3)
```

```python
iris = sns.load_dataset('iris')
iris.tail(4)
```

# 1. Scatterplot (Numerical - Numerical)

### bivariate-analysis

```python
sns.scatterplot(x=tips['total_bill'], y=tips['tip'], hue=tips['sex'])
```

```python
## add more info
sns.scatterplot(x=tips['total_bill'], y=tips['tip'], hue=tips['sex'], style=tips['smoker'])
```

```python
# to see in each team how many meamber was!
## add more info
sns.scatterplot(x=tips['total_bill'], y=tips['tip'], hue=tips['sex'], style=tips['smoker'], size=tips['size'])
```

# 3. Box plot (Numerical - Categorical)

```python
sns.boxplot(data=titanic, x="Sex", y="Age",palette="Set2")
```

```python
sns.boxplot(data=titanic, x="Sex", y="Age", hue="Survived", palette="Set2")
```

# 4. Distplot (Numerical - Categorical)

```python
sns.kdeplot(data=titanic, x = "Age", hue="Survived")
```

# 5. HeatMap (Catg - Catg)

```python
titanic.head(1)
```

```python
pd.crosstab(titanic["Pclass"], titanic["Survived"])
```

```python
sns.heatmap(pd.crosstab(titanic["Pclass"], titanic["Survived"]))
```

```python
# Calculate the mean survival rate for each passenger class
titanic.groupby('Pclass')['Survived'].mean()
```

```python
titanic.groupby('Sex')['Survived'].mean()
```

```python
titanic.groupby('Embarked')['Survived'].mean()
```

# ClusterMap (Catg - Catg)

```python
pd.crosstab(titanic['SibSp'], titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```
