# Univariate Analysis of Titanic datasets

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/MDalamin5/Machine-Learning-2.0/refs/heads/main/2.5-Understanding-Data-EDA/Basics-Question-To-Data/train.csv")
df.sample(5)
```

# Categorical data

### Countplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df['Survived'].value_counts()
```

```python
sns.countplot(df, x=df['Survived'])
```

```python
sns.countplot(df, x=df['Pclass'], palette='Set1')
```

```python
sns.countplot(df, x=df['Embarked'], palette="Set2")
```
